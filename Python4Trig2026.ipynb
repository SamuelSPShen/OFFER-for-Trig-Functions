#### Figure 1: 

All six trigonometric functions can be illustrated by <br> this single figure with triangles associated with a unit circle


```python
import numpy as np
import matplotlib.pyplot as plt

# Define sec and csc functions as Python does not 
# have them in standard libraries
def sec(x):
    return 1 / np.cos(x)

def csc(x):
    return 1 / np.sin(x)

# Unit circle
n = 401
t = np.linspace(0, 2 * np.pi, n)
x = np.cos(t)
y = np.sin(t)

# Specify angle
theta = 53 * np.pi / 180

# Create figure
fig, ax = plt.subplots(figsize=(6.00, 5.2), dpi=600)

# Remove margins / axes for publications
fig.subplots_adjust(left=0, right=1, bottom=0, top=1)
ax.axis("off")
ax.set_aspect("equal", adjustable="box")

# Plot unit circle
ax.plot(x, y, color="black", linewidth=1)

xmax = sec(theta) + 0.2
ymax = csc(theta) + 0.2
ax.set_xlim(-1.2, xmax)
ax.set_ylim(-1.2, ymax)

# Plot x and y axes with arrows
ax.annotate("", xy=(xmax, 0), xytext=(-1.2, 0),
            arrowprops=dict(arrowstyle="->", linewidth=1))
ax.annotate("", xy=(0, ymax), xytext=(0, -1.2),
            arrowprops=dict(arrowstyle="->", linewidth=1))

# Label axes and origin
ax.text(xmax - 0.05, -0.1, r"$x$")
ax.text(-0.10, ymax - 0.05, r"$y$")
ax.text(-0.1, -0.1, "O")

# Mark point A on unit circle
ax.plot(np.cos(theta), np.sin(theta), "ko", markersize=3)

# Plot theta angle arc
t1 = np.linspace(0, theta, n)
rthe = 0.15
x1 = rthe * np.cos(t1)
y1 = rthe * np.sin(t1)
ax.plot(x1, y1, color="black", linewidth=1)

mid = n // 2
ax.text(
    1.5 * rthe * np.cos(t1[mid]),
    1.5 * rthe * np.sin(t1[mid]),
    r"$\theta$"
)

# Plot triangle sides 
ax.plot([0, sec(theta)], [0.01, 0.01], 
        linewidth=2, color="magenta")
ax.plot([np.cos(theta), sec(theta)], [np.sin(theta), 0],
        linewidth=2, color="orange")
ax.plot([np.cos(theta), 0], [np.sin(theta), csc(theta)],
        linewidth=2, color="brown")
ax.plot([0, 0], [0, csc(theta)], linewidth=2.5, color="green")
ax.plot([np.cos(theta), 0], [np.sin(theta), 0],
        linewidth=1, color="black")
ax.plot([np.cos(theta), np.cos(theta)], [np.sin(theta), 0],
        linewidth=2, color="red")
ax.plot([np.cos(theta), np.cos(theta)], [-np.sin(theta), 0],
        linestyle="dotted", linewidth=2, color="red")
ax.plot([np.cos(theta), 0], [np.sin(theta), np.sin(theta)],
        linewidth=1, color="black")
ax.plot([np.cos(theta), 0], [-0.01, -0.01],
        linewidth=2, color="blue")

# Point labels
ax.text(np.cos(theta) + 0.05, np.sin(theta) + 0.02, "A")
ax.text(np.cos(theta) + 0.04, -0.08, "B")
ax.text(sec(theta), -0.08, "C")
ax.text(-0.1, csc(theta), "D")
ax.text(-0.1, np.sin(theta), "E")
ax.text(np.cos(theta), -np.sin(theta) - 0.08, "G")

# Label trig functions
ax.text(np.cos(theta) + 0.05, np.sin(theta) / 2 - 0.1,
        r"$\sin\ \theta$", rotation=90, color="red")

ax.text((np.cos(theta) + sec(theta)) / 2 + 0.05,
        np.sin(theta) / 2 - 0.1,
        r"$\tan\ \theta$",
        rotation=-(90 - theta * 180 / np.pi),
        color="orange")

ax.text(sec(theta) - 0.5, 0.05,
        r"$\sec\ \theta$", color="magenta")

# Complementary trig labels
ax.text(np.cos(theta) / 2.4, -0.1,
        r"$\cos\ \theta$", color="blue")

ax.text(np.cos(theta) / 2,
        (np.sin(theta) + csc(theta)) / 2 - 0.05,
        r"$\cot\ \theta$",
        rotation=-(90 - theta * 180 / np.pi),
        color="brown")

ax.text(-0.23, csc(theta) / 2.5,
        r"$\csc\ \theta$", color="green")

# Radius label
ax.text(np.cos(theta) / 2, np.sin(theta) / 2 + 0.1, "1")

# Unit circle label
ax.text(-np.cos(theta), -np.sin(theta) - 0.1,
        "Unit Circle",
        rotation=270 + theta * 180 / np.pi)

# Save figure
plt.savefig("fig01.png", dpi=600, bbox_inches="tight", pad_inches=0)
plt.close()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define sec and csc functions
def sec(x):
    return 1 / np.cos(x)

def csc(x):
    return 1 / np.sin(x)

# Plot parameters
n = 401
t = np.linspace(0, 2 * np.pi, n)
x = np.cos(t)
y = np.sin(t)

xext = 1.0
yext = 0.2

# Angle
theta = np.deg2rad(53)

# Create figure
fig, ax = plt.subplots(figsize=(6.0, 3.48), dpi=600)

# Remove margins
fig.subplots_adjust(left=0, right=1, bottom=0, top=1)
ax.axis("off")
ax.set_aspect("equal")

# Plot unit circle
ax.plot(x, y, color="black", linewidth=1)

xmin = -1.2 * sec(theta)
xmax = sec(theta) + xext
ymin = -1.2
ymax = csc(theta) + yext

ax.set_xlim(xmin, xmax)
ax.set_ylim(ymin, ymax)

# --------------------------------------------------
# Axes
# --------------------------------------------------
ax.annotate(
    "",
    xy=(xmax, 0),
    xytext=(xmin, 0),
    arrowprops=dict(arrowstyle="->", linewidth=1),
)

ax.annotate(
    "",
    xy=(0, ymax),
    xytext=(0, ymin),
    arrowprops=dict(arrowstyle="->", linewidth=1),
)

#  Label the axes
ax.text(xmax -0.03, -0.12, r"$x$")
ax.text(-0.10, ymax -0.02, r"$y$")
ax.text(-0.10, -0.10, "O")

# --------------------------------------------------
#  Plot points A, H
# --------------------------------------------------
A = (np.cos(theta), np.sin(theta))

ax.plot(A[0], A[1], "ko", markersize=5)
ax.plot(-sec(theta), 0, "o", color="magenta", markersize=3)

# --------------------------------------------------
# Theta arc
# --------------------------------------------------
rthe = 0.15
t1 = np.linspace(0, theta, n)

ax.plot(
    rthe * np.cos(t1),
    rthe * np.sin(t1),
    color="black", linewidth=0.7
)

mid = len(t1) // 2
ax.text(
    1.2 * rthe * np.cos(t1[mid]),
    1.2 * rthe * np.sin(t1[mid]),
    r"$\theta$",
)

# --------------------------------------------------
# Plot triangle segments/sides
# --------------------------------------------------
ax.plot([0, sec(theta)], [0.01, 0.01],
        color="magenta", linewidth=2)

ax.plot([A[0], sec(theta)],
        [A[1], 0],
        color="orange", linewidth=2)

ax.plot([A[0], 0],
        [A[1], csc(theta)],
        color="brown", linewidth=2)

ax.plot([0, 0],
        [0, csc(theta)],
        color="green", linewidth=2)

ax.plot([A[0], 0],
        [A[1], 0],
        color="black", linewidth=1)

ax.plot([A[0], A[0]],
        [A[1], 0],
        color="red", linewidth=2)

ax.plot([A[0], A[0]],
        [-A[1], 0],
        color="red",
        linestyle=":",
        linewidth=2)

ax.plot([A[0], 0],
        [A[1], A[1]],
        color="black",
        linewidth=1)

ax.plot([A[0], 0],
        [-0.01, -0.01],
        color="blue",
        linewidth=2)

# --------------------------------------------------
# Label points A, B, C, D, E, G, H
# --------------------------------------------------
ax.text(A[0] + 0.03, A[1] + 0.05, "A")
ax.text(A[0] + 0.04, -0.10, "B")
ax.text(sec(theta), -0.10, "C")
ax.text(-0.10, csc(theta), "D")
ax.text(-0.10, np.sin(theta) - 0.05, "E")
ax.text(A[0] + 0.01, -A[1] - 0.08, "G")
ax.text(-sec(theta), -0.12, "H")

# --------------------------------------------------
# Label trigonometric functions
# --------------------------------------------------
ax.text(
    A[0] + 0.05,
    A[1] / 2.5,
    r"$\sin\theta$",
    rotation=90,
    color="red",
)

ax.text(
    (A[0] + sec(theta)) / 2,
    A[1] / 2.8,
    r"$\tan\theta$",
    rotation=-(90 - np.degrees(theta)),
    color="orange",
)

ax.text(
    sec(theta) - 0.60,
    0.06,
    r"$\sec\theta$",
    color="magenta",
)

ax.text(
    A[0] / 2.7,
    -0.12,
    r"$\cos\theta$",
    color="blue",
)

ax.text(
    A[0] / 2,
    (A[1] + csc(theta)) / 2.2,
    r"$\cot\theta$",
    rotation=-(90 - np.degrees(theta)),
    color="brown",
)

ax.text(
    -0.12,
    csc(theta) / 3.2,
    r"$\csc\theta$",
    color="green",
    rotation=90,
)

# Radius label "1" for the unit circle
ax.text(
    A[0] / 2,
    A[1] / 2 + 0.10,
    "1",
)

# --------------------------------------------------
# Non-tangent line
# --------------------------------------------------
Fx = sec(theta) + 0.8 * xext

ax.plot(
    [A[0], Fx],
    [A[1], 0],
    linestyle="--",
    color="black",
    linewidth=1,
)

ax.text(Fx, -0.10, "F")
ax.text(sec(theta), 0.5 * np.sin(theta), "c")

# --------------------------------------------------
# Label the unit circle 
# --------------------------------------------------
ax.text(
    -A[0],
    -A[1] - 0.17,
    "Unit Circle",
    rotation=280 + np.degrees(theta),
)

# Save figure
plt.savefig(
    "fig03.png",
    dpi=600,
    bbox_inches="tight",
    pad_inches=0,
)

plt.close()
```

#### Figure 4: 

A right triangle to support the Pythagorean theorem

```python
import numpy as np
import matplotlib.pyplot as plt

# --------------------------------------------------
# Python code for plotting Fig. 4
# Fig. 4: A right triangle supporting the
# Pythagorean theorem
# --------------------------------------------------

# Parameters
n = 401
theta = np.deg2rad(53)

# Create figure
fig, ax = plt.subplots(figsize=(6.0, 5.2), dpi=600)

# Remove margins
fig.subplots_adjust(left=0, right=1, bottom=0, top=1)

# Set plot limits
ax.set_xlim(-0.2, np.cos(theta) + 0.2)
ax.set_ylim(-0.2, np.sin(theta) + 0.2)

# Equal scaling
ax.set_aspect("equal")

# Remove axes and frame
ax.axis("off")

# --------------------------------------------------
# Theta angle arc
# --------------------------------------------------
rthe = 0.15
t1 = np.linspace(0, theta, n)

x1 = rthe * np.cos(t1)
y1 = rthe * np.sin(t1)

ax.plot(x1, y1, color="black", linewidth=1)

mid = len(t1) // 2
ax.text(
    1.2 * rthe * np.cos(t1[mid]),
    1.2 * rthe * np.sin(t1[mid]),
    r"$\theta$",
    fontsize=14,
)

# --------------------------------------------------
# Plot triangle ABC
# --------------------------------------------------
A = (np.cos(theta), np.sin(theta))
B = (0, 0)
C = (np.cos(theta), 0)

# Hypotenuse
ax.plot(
    [A[0], B[0]],
    [A[1], B[1]],
    color="black",
    linewidth=2,
)

# Vertical side
ax.plot(
    [A[0], C[0]],
    [A[1], C[1]],
    color="red",
    linewidth=2,
)

# Horizontal side
ax.plot(
    [C[0], B[0]],
    [C[1], B[1]],
    color="blue",
    linewidth=2,
)

# --------------------------------------------------
# Label the vertices A, B, C
# --------------------------------------------------
ax.text(
    A[0] + 0.02,
    A[1] + 0.02,
    "A",
    fontsize=13,
)

ax.text(-0.04, -0.03, "B", fontsize=13)

ax.text(
    C[0] + 0.02,
    -0.03,
    "C",
    fontsize=13,
)

# --------------------------------------------------
# Label the sides a, b, c
# --------------------------------------------------

ax.text(
    A[0] / 2,
    -0.04,
    "a",
    fontsize=13,
    color="blue",
)

ax.text(
    A[0] + 0.02,
    A[1] / 2,
    "b",
    fontsize=13,
    color="red",
)

ax.text(
    A[0] / 2,
    A[1] / 2 + 0.07,
    "c",
    fontsize=13,
    color="black",
)

# --------------------------------------------------
# Save figure
# --------------------------------------------------
plt.savefig(
    "fig04.png",
    dpi=600,
    bbox_inches="tight",
    pad_inches=0,
)

plt.close()
```

#### Figure 5: 

A figure for the derivation of the cosine difference formula


```python
import numpy as np
import matplotlib.pyplot as plt

# Create figure 
fig, ax = plt.subplots(figsize=(5.0, 5.0), dpi=600)

# Remove margins
fig.subplots_adjust(left=0, right=1, bottom=0, top=1)

# Set up figure frame
ax.set_xlim(-1.2, 1.2)
ax.set_ylim(-1.2, 1.2)
ax.set_aspect("equal")
ax.axis("off")

# Unit circle
n = 401
t = np.linspace(0, 2 * np.pi, n)
x = np.cos(t)
y = np.sin(t)

ax.plot(x, y, color="black", linewidth=1)

# x- and y-axes with arrows
ax.annotate(
    "",
    xy=(1.2, 0),
    xytext=(-1.2, 0),
    arrowprops=dict(arrowstyle="->", linewidth=1),
)

ax.annotate(
    "",
    xy=(0, 1.2),
    xytext=(0, -1.2),
    arrowprops=dict(arrowstyle="->", linewidth=1),
)

# Axis labels and origin
ax.text(1.13, -0.07, r"$x$")
ax.text(-0.07, 1.13, r"$y$")
ax.text(-0.07, -0.07, "O")

# Angles
alpha = np.deg2rad(70)
beta = np.deg2rad(20)

# Mark points for alpha and beta
ax.plot(np.cos(alpha), np.sin(alpha), "o", color="blue", markersize=4)
ax.plot(np.cos(beta), np.sin(beta), "o", color="red", markersize=4)

# --------------------------------------------------
# Alpha angle arc
# --------------------------------------------------
t1 = np.linspace(0, alpha, n)
rthe = 0.15

ax.plot(
    rthe * np.cos(t1),
    rthe * np.sin(t1),
    color="blue",
    linewidth = 0.7,
)

mid = len(t1) // 2
ax.text(
    1.01 * rthe * np.cos(t1[mid]),
    1.35 * rthe * np.sin(t1[mid]),
    r"$\alpha$",
    color="blue",
)

# --------------------------------------------------
# Beta angle arc
# --------------------------------------------------
t1 = np.linspace(0, beta, n)
rthe = 0.45

ax.plot(
    rthe * np.cos(t1),
    rthe * np.sin(t1),
    color="red",
    linewidth = 0.7,
)

mid = len(t1) // 2
ax.text(
    1.04 * rthe * np.cos(t1[mid]),
    1.05 * rthe * np.sin(t1[mid]),
    r"$\beta$",
    color="red",
)

# --------------------------------------------------
# Alpha - beta angle arc
# --------------------------------------------------
t1 = np.linspace(beta, alpha, n)
rthe = 0.27

ax.plot(
    rthe * np.cos(t1),
    rthe * np.sin(t1),
    color="brown",
    linewidth=0.7,
)

mid = len(t1) // 2
ax.text(
    0.65 * rthe * np.cos(t1[mid]),
    0.9 * rthe * np.sin(t1[mid]),
    r"$\alpha - \beta$",
    rotation=-45,
    color="brown",
)

# --------------------------------------------------
# Triangle and projection sides
# --------------------------------------------------

# Alpha triangle
ax.plot(
    [0, np.cos(alpha)],
    [0, np.sin(alpha)],
    linewidth=2,
    color="blue",
)

ax.plot(
    [np.cos(alpha), np.cos(alpha)],
    [0, np.sin(alpha)],
    linewidth=2,
    color="blue",
)

ax.plot(
    [0, np.cos(alpha)],
    [0.01, 0.01],
    linewidth=2,
    color="blue",
)

# Beta triangle
ax.plot(
    [0, np.cos(beta)],
    [0, np.sin(beta)],
    linewidth=3,
    color="red",
)

ax.plot(
    [np.cos(beta), np.cos(beta)],
    [0, np.sin(beta)],
    linewidth=3,
    color="red",
)

ax.plot(
    [0, np.cos(beta)],
    [-0.01, -0.01],
    linewidth=3,
    color="red",
)

# Dashed comparison segments
ax.plot(
    [np.cos(alpha), np.cos(beta)],
    [np.sin(beta), np.sin(beta)],
    linestyle=":",
    linewidth=1.2,
    color="black",
)

ax.plot(
    [np.cos(alpha), np.cos(beta)],
    [np.sin(alpha), np.sin(beta)],
    linestyle=":",
    linewidth=1.2,
    color="black",
)

# --------------------------------------------------
# Label points A, B, C, D, E
# --------------------------------------------------
ax.text(np.cos(alpha) + 0.01, np.sin(alpha) + 0.03, "A", color="blue")
ax.text(np.cos(beta) + 0.03, np.sin(beta) + 0.01, "B", color="red")
ax.text(np.cos(alpha) -0.01, -0.09, "C", color="blue")
ax.text(np.cos(beta) - 0.01, -0.09, "D", color="red")
ax.text(np.cos(alpha) - 0.07, np.sin(beta), "E", color="blue")

# --------------------------------------------------
# Label sin and cos 
# --------------------------------------------------
ax.text(
    np.cos(alpha) + 0.04,
    np.sin(alpha) / 1.9,
    r"$\sin\alpha$",
    rotation=90,
    color="blue",
)

ax.text(
    np.cos(beta) - 0.11,
    np.sin(beta) / 2.8,
    r"$\sin\beta$",
    rotation=90,
    color="red",
)

ax.text(
    np.cos(alpha) / 3.1,
    -0.08,
    r"$\cos\alpha$",
    color="blue",
)

ax.text(
    np.cos(beta) / 1.8,
    -0.08,
    r"$\cos\beta$",
    color="red",
)

# Label the unit circle label

ax.text(
    -np.cos(2*alpha),
    -np.sin(2*alpha) - 0.1,
    "Unit Circle",
    rotation=276 + np.degrees(2*alpha),
)

# Save figure
plt.savefig(
    "fig05.png",
    dpi=600,
    bbox_inches="tight",
    pad_inches=0,
)

plt.close()
```

#### Figure 6: 

Right triangles with different lengths of hypotenuse

```python
import numpy as np
import matplotlib.pyplot as plt

# Define sec function
def sec(x):
    return 1 / np.cos(x)

# Parameters
theta = np.deg2rad(53)

n = 101
t = np.linspace(0, theta, n)

x = np.cos(t)
y = np.sin(t)

r = 0.2
xext = 0.2
yext = 0.2

# Create figure
fig, ax = plt.subplots(figsize=(5.0, 3.5), dpi=600)

# Remove margins
fig.subplots_adjust(left=0, right=1, bottom=0, top=1)

# Set plot limits
ax.set_xlim(-xext, sec(theta) + xext)
ax.set_ylim(-yext, 1 / np.sin(theta) + 1.2 * yext)

ax.set_aspect("equal")
ax.axis("off")

# --------------------------------------------------
# Plot theta arc
# --------------------------------------------------
ax.plot(r * x, r * y, color="black", linewidth=1)

mid = len(t) // 2
ax.text(
    1.2 * r * x[mid],
    1.2 * r * y[mid],
    r"$\theta$",
)

# Origin
ax.text(-0.08, -0.08, "O", color="magenta")
ax.plot(0, 0, "o", color="blue", markersize=5)

# --------------------------------------------------
# Larger triangle
# --------------------------------------------------
R = 1.8

Dx = R * np.cos(theta)
Dy = R * np.sin(theta)

# Hypotenuse
ax.plot(
    [0, Dx],
    [0, Dy],
    linewidth=4,
    color="magenta",
)

ax.plot(Dx, Dy, "o", color="magenta", markersize=4)
ax.text(Dx, Dy + 0.05, "D", color="magenta")

# Base
ax.plot(
    [0, Dx],
    [0, 0],
    linewidth=4,
    color="magenta",
)

ax.plot(Dx, 0, "o", color="magenta", markersize=4)
ax.text(Dx, -0.08, "C", color="magenta")

# Vertical side
ax.plot(
    [Dx, Dx],
    [0, Dy],
    linewidth=4,
    color="magenta",
)

# --------------------------------------------------
# Unit triangle
# --------------------------------------------------
Ax = np.cos(theta)
Ay = np.sin(theta)

# Hypotenuse
ax.plot(
    [0, Ax],
    [0, Ay],
    linewidth=2,
    color="black",
)

ax.plot(Ax, Ay, "ko", markersize=4)
ax.text(Ax - 0.05, Ay + 0.04, "A")

# Base
ax.plot(
    [0, Ax],
    [0, 0],
    linewidth=2,
    color="black",
)

# Vertical side
ax.plot(
    [Ax, Ax],
    [0, Ay],
    linewidth=2,
    color="black",
)

ax.plot(Ax, 0, "ko", markersize=4)
ax.text(Ax, -0.08, "B")

# --------------------------------------------------
# Side labels (large triangle)
# --------------------------------------------------
ax.text(
    0.7 * Dx,
    -0.07,
    "a",
    color="magenta",
)

ax.text(
    0.65 * Dx,
    0.65 * Dy + 0.09,
    "c",
    color="magenta",
)

ax.text(
    Dx + 0.04,
    0.5 * Dy,
    "b",
    color="magenta",
)

# --------------------------------------------------
# Trigonometric labels (unit triangle)
# --------------------------------------------------
ax.text(
    0.45 * Ax,
    -0.07,
    r"$\cos\theta$",
)

ax.text(
    Ax + 0.04,
    0.45 * Ay,
    r"$\sin\theta$",
    rotation=90,
)

# --------------------------------------------------
# Save figure
# --------------------------------------------------
plt.savefig(
    "fig06.png",
    dpi=600,
    bbox_inches="tight",
    pad_inches=0,
)

plt.close()
```

A figure for deriving the derivative formula for sine theta


```python
import numpy as np
import matplotlib.pyplot as plt

# Create figure
fig, ax = plt.subplots(figsize=(6.0, 6.0), dpi=600)
fig.subplots_adjust(left=0, right=1, bottom=0, top=1)

# Figure frame
ax.set_xlim(-1.2, 1.2)
ax.set_ylim(-1.2, 1.2)
ax.set_aspect("equal")
ax.axis("off")

# Unit circle
n = 401
t = np.linspace(0, 2 * np.pi, n)
x = np.cos(t)
y = np.sin(t)
ax.plot(x, y, color="black", linewidth=1)

# Axes
ax.annotate(
    "",
    xy=(1.2, 0),
    xytext=(-1.2, 0),
    arrowprops=dict(arrowstyle="->", linewidth=1),
)

ax.annotate(
    "",
    xy=(0, 1.2),
    xytext=(0, -1.2),
    arrowprops=dict(arrowstyle="->", linewidth=1),
)

# Label axes and origin
ax.text(1.15, -0.07, r"$x$")
ax.text(-0.08, 1.15, r"$y$")
ax.text(-0.08, -0.08, "O", fontsize = 14)

# Angles
theta = np.deg2rad(35)
Delthe = np.deg2rad(15)
thet2 = theta + Delthe

# Mark points for theta and theta + phi
ax.plot(np.cos(theta), np.sin(theta), "o", color="red", markersize=3)
ax.plot(np.cos(thet2), np.sin(thet2), "o", color="blue", markersize=3)

# Theta arc
rthe = 0.15
t1 = np.linspace(0, theta, n)
ax.plot(rthe * np.cos(t1), rthe * np.sin(t1), 
        color="red", linewidth = 0.7)

mid = len(t1) // 2
ax.text(
    1.1 * rthe * np.cos(t1[mid]),
    0.9 * rthe * np.sin(t1[mid]),
    r"$\theta$",
    color="red",
)

# Increment angle phi arc
rthe = 0.45
t1 = np.linspace(theta, thet2, n)
ax.plot(rthe * np.cos(t1), rthe * np.sin(t1), 
        color="blue", linewidth = 0.7)

mid = len(t1) // 2
ax.text(
    1.05 * rthe * np.cos(t1[mid]),
    1.10 * rthe * np.sin(t1[mid]),
    r"$\phi$",
    color="blue",
)

# Main red triangle
ax.plot(
    [0, np.cos(theta)],
    [0, np.sin(theta)],
    linewidth=2,
    color="red",
)

ax.plot(
    [np.cos(theta), np.cos(theta)],
    [0, np.sin(theta)],
    linewidth=2,
    color="red",
)

ax.plot(
    [0, np.cos(theta)],
    [0, 0],
    linewidth=2,
    color="red",
)

# Blue radius and projections
ax.plot(
    [0, np.cos(thet2)],
    [0, np.sin(thet2)],
    linewidth=1,
    color="blue",
)

ax.plot(
    [0, np.cos(thet2)],
    [0, 0],
    linewidth=1,
    color="blue",
)

ax.plot(
    [np.cos(thet2), np.cos(theta)],
    [np.sin(theta), np.sin(theta)],
    linewidth=1,
    color="black",
)

ax.plot(
    [np.cos(thet2), np.cos(thet2)],
    [0, np.sin(thet2) + 0.3],
    linewidth=1,
    color="blue",
)

# Tangent line at theta
tt = np.linspace(-0.7, 0.7, n)
ax.plot(
    np.cos(theta) + tt * np.sin(theta),
    np.sin(theta) - tt * np.cos(theta),
    linewidth=1,
    color="magenta",
)

# Label points A, B, C, D, E, F
ax.text(np.cos(theta) + 0.01, np.sin(theta) + 0.03, "A", color="red")
ax.text(np.cos(theta), -0.07, "B", color="red")
ax.text(np.cos(thet2) - 0.08, np.sin(thet2) - 0.03, "C", color="blue")
ax.text(np.cos(thet2) + 0.03, np.sin(thet2) + 0.05, "D", color="magenta")
ax.text(np.cos(thet2) - 0.06, np.sin(theta), "E", color="blue")
ax.text(np.cos(thet2), -0.07, "F", color="red")

ax.text(
    0.65 * np.cos(theta) - 0.03,
    0.65 * np.sin(theta) + 0.03,
    "1",
    color="red",
)

# Magenta point on tangent line
magenta_y = np.sin(theta) - np.tan(theta - np.pi / 2) * (
    np.cos(theta) - np.cos(thet2)
)

ax.plot(
    np.cos(thet2),
    magenta_y,
    "o",
    color="magenta",
    markersize=3,
)

# Label sin and cos functions
ax.text(
    np.cos(theta) + 0.03,
    np.sin(theta) / 2.2,
    r"$\sin\theta$",
    rotation=90,
    color="red",
)

ax.text(
    0.50 * np.cos(theta),
    -0.06,
    r"$\cos\theta$",
    color="red",
)

ax.text(
    0.40 * np.cos(thet2),
    0.03,
    r"$\cos(\theta+\phi)$",
    color="blue",
)

ax.text(
    np.cos(thet2) + 0.03,
    np.sin(thet2) / 5,
    r"$\sin(\theta+\phi)$",
    rotation=90,
    color="blue",
)

# Label the unit circle
ax.text(
    np.cos(np.pi / 4),
    -np.sin(np.pi / 4) - 0.1,
    "Unit Circle",
    rotation=49,
)

# Save figure
plt.savefig("fig07.png", dpi=600, bbox_inches="tight", pad_inches=0)
plt.close()
```
